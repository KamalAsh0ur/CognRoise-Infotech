```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load data

```python
test = pd.read_csv('/content/Test.csv', encoding='latin1')
train = pd.read_csv('/content/Train.csv', encoding='latin1')
```

# Clean Data

```python
test.dropna(inplace=True)
test
```

```python
train.dropna(inplace=True)
train
```

# combine -->> append queries

```python

Dataset = pd.concat([train, test], ignore_index=True)
Dataset.dropna(inplace=True)
Dataset
```

```python
from google.colab import sheets
sheet = sheets.InteractiveSheet(df=Dataset)
```

```python
Dataset.describe()
```

```python

Sales = Dataset.groupby('Item_Type')['Item_Outlet_Sales'].sum()
Sales = Sales.to_frame()
```

# Analysis 📊

```python
Dataset.plot(kind='scatter', x='Item_Weight', y='Item_Visibility', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python

Dataset.groupby('Item_Fat_Content').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python

Dataset.groupby('Outlet_Size').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
Sales
```

```python
Sales.sort_values(by='Item_Outlet_Sales', ascending=True).plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python

Sales = Dataset.groupby('Outlet_Type')['Item_Outlet_Sales'].sum()
Sales = Sales.to_frame()
Sales.sort_values(by='Item_Outlet_Sales', ascending=False).plot(kind='pie', subplots=True, figsize=(15, 8), autopct='%1.1f%%', shadow=True)
plt.gca().spines[['top', 'right',]].set_visible(False)
```
